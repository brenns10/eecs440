```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
x = np.arange(0, 6, 0.01)
```

```python
y = np.arange(0, 5, 0.01)
```

```python
y1 = 9 - x
```

```python
y2 = 0.5 * x
```

```python
y3 = x + 3 
```

```python
fig, ax = plt.subplots()
```

```python
ax.set_xlim(-1, 7)
ax.set_ylim(-1, 6)
```

```python
ax.plot(x, y1, color='b')
ax.plot(x, y2, color='g')
ax.plot(x, y3, color='r')
```

```python
fig
```

```python
ax.vlines(0, -1, 6)
ax.hlines(0, -1, 7)
ax.hlines(5, -1, 7, colors='c')
```

```python
fig
```

```python
sh_x_1 = np.arange(0, 2, 0.01)
sh_x_2 = np.arange(2, 4, 0.01)
sh_x_3 = np.arange(4, 6, 0.01)
sh_y_1 = sh_x_1 + 3
sh_y_2 = np.array([5] * 200)
sh_y_3 = 9 - sh_x_3
```

```python
sh_x = np.append(np.append(sh_x_1, sh_x_2), sh_x_3)
sh_y = np.append(np.append(sh_y_1, sh_y_2), sh_y_3)
sh_bot = 0.5 * sh_x
```

```python
ax.fill_between(sh_x, sh_y, sh_bot)
```

```python
fig
```

```python
fig.savefig('feasible.pdf')
```

```python
ax.set_title('Feasible Region')
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
```
